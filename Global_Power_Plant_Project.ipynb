### Let's remove irrelevant and empty columns, before imputing values.

```python
df.drop(columns=["country_long",'url','year_of_capacity_data',
                 'gppd_idnr','owner','other_fuel3','other_fuel2',
                 'estimated_generation_gwh','wepp_id'],axis=1,inplace=True)
```

```python
missing_values = df.isnull().sum()
(missing_values/len(df))*100
```

1. Let's impute latitude and longitude with mean or median as mean wil, be same at the end.
2. In generation_data_source nearly 44% data is missing, it does not have much importance in our analysis, lets drop it.
3.  As commissioning_year for most of industrial powerplant is missing(42%), after checking correlation we will decide to keep or drop this features.
4. We have 7 different columns of generation_gwh for year 2013-2019:
   It is important feature for important plants.
   We cannot do any imputation by mean or median since different power plants have different generation capacity.
   Therefore we will keep the feature along with missing values and perform further analysis.
5. other_fuel1 is also an important feature:
    But in case of renewable energy source other fuel is not important since they will operate only on specific sources.
    For renewable energy source no any other fuel is needed, so we can impute them with 'Not Applicable'. At the end we are going to encode them.

```python
# Missing Value Imputation of longitude with median
df['longitude'] = df['longitude'].fillna(df['longitude'].median())
```

```python
# Missing Value Imputation of latitude with mean
df['latitude'] = df['latitude'].fillna(df['latitude'].mean())
```

```python
df.drop(['country'],axis=1,inplace=True)
```
